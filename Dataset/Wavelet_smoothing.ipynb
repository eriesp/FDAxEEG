# Wavelet smoothing

```python
import numpy as np
import mne
import glob
import os
import matplotlib as plt
import pywt
```

```python
directory = os.getcwd()

path_adhd = "\ICA_filtered\ADHD"
matfiles_adhd = glob.glob(directory + path_adhd + '/*.edf')

#path_control = "\Control"
#matfiles_control = glob.glob(directory + path_control + '/*.mat')
```

```python
matfiles_adhd
```

```python
raw = mne.io.read_raw_edf(matfiles_adhd[0])
```

```python
signals = raw.get_data()
```

```python
fp1=signals[0]
```

```python
raw.plot()
```

```python
coeffs = pywt.wavedec(fp1, 'db2', level=4)
coeffs
```

```python
A=coeffs[1]
```

```python
plt.pyplot.plot(A)
plt.pyplot.show()
```
